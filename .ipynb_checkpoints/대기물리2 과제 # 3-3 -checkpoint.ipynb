```python
# 필요 모듈 import 
import datetime
import numpy as np
import matplotlib.pyplot as plt

# Coefficient 정의
an=[1.000110, 0.034221, 0.000719]
bn=[0, 0.001280,0.000077]
cn=[0.006918, -0.399912, -0.006758, -0.002697]
dn=[0, 0.070257, 0.000907, 0.000148]
S = 1367    # 태양상수
lat = 37.57*np.pi/180   # 연세대학교 위도(rad으로 변환)
t = 2*np.pi*325/365 # 11월 22일 day number
gap = np.pi/12   # Time gap

# 11월 22일 (a/r)^2과 delta값 구하기 (t = 325)
AR = 0
for j in range(0, 3):
    AR = AR + an[j] * np.cos(j * t) + bn[j] * np.sin(j * t) # (a/r)^2
delta = 0
for k in range(0, 4):
    delta = delta + cn[k] * np.cos(k * t) + dn[k] * np.sin(k * t) # delta

# Q_hourly 구하고 print
print('Time      Hourly solar insolation(Q)[W/m2]')
print('----      --------------------------------')

for j in range(0, 9):
    Q_hourly[j] = S*AR*(np.sin(lat)*np.sin(delta)+(24/np.pi)*np.cos(lat)*np.cos(delta)*np.sin(np.pi/24)*np.cos((j - 4)*gap))
    print(8+j,':17:50  ',  Q_hourly[j])

# local noon 정의
LN = datetime.datetime(2022,11,22,17,50) # local noon
datetime.datetime(2022,11,22,17,50)

# plotting 
plt.figure()
plt.plot(Q_hourly)
plt.title('Hourly Solar Insolation at 11/22, Yonsei University')
plt.xlabel('Time')
plt.ylabel('Solar Insolation(Q) [W/m^2]')
plt.xticks([0,1,2,3,4,5,6,7,8],labels = ['LN-h4','LN-h3','LN-h2','LN-h1','LN','LN+h1','LN+h2','LN+h3','LN+h4'])
plt.show()
```
