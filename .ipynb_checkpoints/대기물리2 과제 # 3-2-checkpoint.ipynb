```python
# 계산을 위해 사용할 모듈 선언
import numpy as np
import matplotlib.pyplot as plt
import datetime
import pandas as pd

# 아래 함수에 넣을 t 구하기
time=[]
for i in range(0,365):
    time.append(2*np.pi*i/365) # i=0 for Jan.1 
                               # i=364 for Dec.31
# 필요한 상수들 정의하기
an = [1.000110,0.034221,0.000719] # an값
bn = [0,0.001280,0.000077] # bn값
cn = [0.006918,-0.399912,-0.006758,-0.002697] # cn값
dn = [0,0.070257,0.000907,0.000148] # dn값
S = 1367 # 태양 상수
lat = 37.57 # 위도
long = 126.97 # 경도

# (a/r)의 제곱 근사하는 함수
def ar(t):
    X=0
    for i in range (0,3):
        X+=an[i]*np.cos(i*t)+bn[i]*np.sin(i*t)
    return X

# (a/r)의 제곱 구하기
AR=[]
for i in range(0, 365):
    AR.append(ar(time[i]))

# decl_angle 근사하는 함수
def decl_angle(t):
    X=0
    for i in range(0,4):
        X+=cn[i]*np.cos(i*t)+dn[i]*np.sin(i*t)
    return X

# decl_angle 구하기
Decl_angle=[]
for i in range(0, 365):
    Decl_angle.append(decl_angle(time[i]))
    
# Half day length 구하는 함수
def hdl(declination):
    X = np.arccos(-np.tan(lat*np.pi/180)*np.tan(declination))
    return X

# Half day length 구하기
Hdl = []
for i in range(0,365):
    Hdl.append(hdl(Decl_angle[i]))
    
# Q 구하기
Q=[]
for i in range(0,365):
    Q.append((S/np.pi)*AR[i]*(np.sin(lat*np.pi/180)*np.sin(Decl_angle[i])*Hdl[i]+np.cos(lat*np.pi/180)*np.cos(Decl_angle[i])*np.sin(Hdl[i])))

# 표 만들기
start_day = datetime.date(2022,1,1)
end_day = datetime.date(2022,12,31)
number_of_days = (end_day-start_day).days+1 

date_arr = []
for i in range(0,365):
    date_arr.append(start_day+datetime.timedelta(days=i))
date_arr = np.array(date_arr)

df = pd.DataFrame({'Date': date_arr, 'Solar insolation(Q) [{w/{m}^2}]': Q})
df
```

```python
# plot 그리기
x = np.arange(0,365)
plt.plot(x,Q)
plt.xlabel('Date    (0=JAN 1, 364=DEC 31)')
plt.ylabel('Solar insolation (Q) [W/m2]')
plt.title("Seung-Ok")
plt.show()
```
