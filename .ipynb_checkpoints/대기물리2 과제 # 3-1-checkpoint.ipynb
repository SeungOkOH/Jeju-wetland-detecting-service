```python
# 계산을 위해 사용할 모듈 선언
import numpy as np
import matplotlib.pyplot as plt
import datetime

# 필요한 배열 선언
latitude = np.arange(0,90+1,15)
delta = np.array([0, 23.5, 0, -23.5])
H = np.zeros((latitude.shape[0], delta.shape[0]))

# H 계산
for i in range(H.shape[0]): # latitude의 shape
    for j in range(H.shape[1]): # delta의 shape
        if delta[j]>0 and 90-abs(delta[j])<latitude[i]: # 백야 case
            H[i, j] = 12
        elif delta[j]<0 and 90-abs(delta[j])<latitude[i]: # 극야 case
            H[i, j] = 0
        else:
            H[i,j] = 12*(np.arccos(-np.tan(latitude[i]*np.pi/180)*np.tan(delta[j]*np.pi/180)))/np.pi
print(H)

# plot 그리기
fig = plt.figure(figsize = (8,6), facecolor = 'white')
ax = fig.add_subplot(1,1,1)
delta_str = ["Spring equinox", "Summer solstice", "Fall equinox", "Winter solstice"] 
for i in range(4):
    ax.plot(latitude, H[:,i], label = '{}'.format(delta_str[i]))
ax.set_yticks(np.linspace(0,12,7))
ax.set_xlabel('Latitude', fontsize = 15)
ax.set_ylabel('Half day length', fontsize = 15)
plt.legend()
```

```python
# 계산을 위해 사용할 모듈 선언
import numpy as np
import matplotlib.pyplot as plt
import datetime
import pandas as pd
import datetime

# 아래 함수에 넣을 t 구하기
time=[]
for i in range(0,365):
    time.append(2*np.pi*i/365) # i=0 for Jan.1 
                               # i=364 for Dec.31
# 필요한 상수들 정의하기
an = [1.000110,0.034221,0.000719] # an값
bn = [0,0.001280,0.000077] # bn값
cn = [0.006918,-0.399912,-0.006758,-0.002697] # cn값
dn = [0,0.070257,0.000907,0.000148] # dn값
S = 1367 # 태양 상수
lat = 37.57 # 위도
long = 126.97 # 경도

# (a/r)의 제곱 근사하는 함수
def ar(t):
    X=0
    for i in range (0,3):
        X+=an[i]*np.cos(i*t)+bn[i]*np.sin(i*t)
    return X

# (a/r)의 제곱 구하기
AR=[]
for i in range(0, 365):
    AR.append(ar(time[i]))

# decl_angle 근사하는 함수
def decl_angle(t):
    X=0
    for i in range(0,4):
        X+=cn[i]*np.cos(i*t)+dn[i]*np.sin(i*t)
    return X

# decl_angle 구하기
Decl_angle=[]
for i in range(0, 365):
    Decl_angle.append(decl_angle(time[i]))
    
# Half day length 구하는 함수
def hdl(declination):
    X = np.arccos(-np.tan(lat*np.pi/180)*np.tan(declination))
    return X

# Half day length 구하기
Hdl = []
for i in range(0,365):
    Hdl.append(hdl(Decl_angle[i]))
    
# Q 구하기
Q=[]
for i in range(0,365):
    Q.append((S/np.pi)*AR[i]*(np.sin(lat*np.pi/180)*np.sin(Decl_angle[i])*Hdl[i]+np.cos(lat*np.pi/180)*np.cos(Decl_angle[i])*np.sin(Hdl[i])))

# 표 만들기
start_day = datetime.date(2022,1,1)
end_day = datetime.date(2022,12,31)
number_of_days = (end_day-start_day).days+1 

date_arr = []
for i in range(0,365):
    date_arr.append(start_day+datetime.timedelta(days=i))
date_arr = np.array(date_arr)

df = pd.DataFrame({'Date': date_arr, 'Solar insolation(Q) [{w/{m}^2}]': Q})
df

# plot 그리기
x = np.arange(0,365)
plt.plot(x,Q)
plt.xlabel('Date    (0=JAN 1, 364=DEC 31)')
plt.ylabel('Solar insolation (Q) [W/m2]')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

an=[1.000110,0.034221,0.000719] #an값
bn=[0,0.001280,0.000077] #bn값
cn=[0.006918,-0.399912,-0.006758,-0.002697] #cn값
dn=[0,0.070257,0.000907,0.000148] #dn값
S=1367 #태양 상수
P=37.57 #위도

#(a/r)의 제곱 구하는 함수
def ar(T):
    X=0
    for i in range (0,3):
        X+=an[i]*np.cos(i*T)+bn[i]*np.sin(i*T)
    return X

#델타 구하는 함수
def delta(T):
    X=0
    for i in range(0,4):
        X+=cn[i]*np.cos(i*T)+dn[i]*np.sin(i*T)
    return X
    
#Half day length 구하는 함수
def HDL(inclination):
    X=np.arccos(-np.tan(P*np.pi/180)*np.tan(inclination))
    return X
    
#함수에 넣을 T 구하기
time=[]
for i in range(0,365):
    time.append(2*np.pi*i/365)

#(a/r)의 제곱 구하기
AR=[]
for i in range(len(time)):
    AR.append(ar(time[i]))

#델타 구하기
DT=[]
for i in range(len(time)):
    DT.append(delta(time[i]))

#Half day length 구하기
Half=[]
for i in range(len(time)):
    Half.append(HDL(DT[i]))

#Q 구하기
Q=[]
for i in range(len(time)):
    Q.append((S/np.pi)*AR[i]*(np.sin(P*np.pi/180)*np.sin(DT[i])*Half[i]+np.cos(P*np.pi/180)*np.cos(DT[i])*np.sin(Half[i])))

#pop 명령어를 사용할 것이기 때문에 plot용 리스트를 하나 더 생성
Qp=[]
for i in range(len(Q)):
    Qp.append(Q[i])
             
#List의 pop 명령어를 사용하여 값을 뽑아냄과 동시에 리스트에서 제거
#이후 format에 맞춰 출력함
Month=[32,29,32,31,32,31,32,32,31,32,31,32]
Monthstr=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','0CT','N0V','DEC']
print(format('Date','^6'),end='')
print('     ',end='')
print(format('Solar insolation (Q) [W/m2]','^27'))
print('-'*6,'     ','-'*27,sep='')
#출력 시퀀스
for i in range(len(Month)):
    for j in range(1,Month[i]):
        print(Monthstr[i],end=' ')
        print(format(j,'>2'),end='')
        print('     ',end='')
        print(format(Q.pop(0),'<27'))

#Plotting
x=np.arange(0,365)
plt.plot(x,Qp)
plt.xlabel('Date    (0=JAN 1, 364=DEC 31)')
plt.ylabel('Solar insolation (Q) [W/m2]')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

month=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT', 'NOV','DEC']
Date=np.array([])
for n1 in range(1,13,1):
    if n1==2:
        for n11 in range(0,28,1):
            Date=np.append(Date, (month[n1-1], n11+1), axis=0)
    if n1==1 or n1==3 or n1==5 or n1==7 or n1==8 or n1==10 or n1==12:
        for n12 in range(0,31,1):
            Date=np.append(Date, (month[n1-1], n12+1), axis=0)
    if n1==4 or n1==6 or n1==9 or n1==11:
        for n13 in range(0,30,1):
            Date=np.append(Date, (month[n1-1], n13+1), axis=0)

an=[1.000110, 0.034221, 0.000719]
bn=[0, 0.001280,0.000077]
cn=[0.006918, -0.399912, -0.006758, -0.002697]
dn=[0, 0.070257, 0.000907, 0.000148]
insol=np.array([])
print('Date   Solar insolation(Q)[W/m2]')
print('-----  ---------------------')
for i in range(0,365,1):
    t=2*np.pi*i/365
    d=0
    for k in range(0,4,1):
        d=d+cn[k]*np.cos(k*t)+dn[k]*np.sin(k*t)
    lat=2*np.pi*(37.57)/360
    cosH=-np.tan(lat)*np.tan(d)
    H=np.arccos(cosH)
    ra=0
    for k2 in range(0,3,1):
        ra=ra+an[k2]*np.cos(k2*t)+bn[k2]*np.sin(k2*t)
    Q= 1367*ra*(H*np.sin(lat)*np.sin(d)+np.cos(lat)*np.cos(d)*np.sin(H))/np.pi
    print(Date[2*i], Date[2*i+1], Q)
    insol=np.append(insol,[float(Q)], axis=0)

plt.plot(insol)
plt.show()



    


```
