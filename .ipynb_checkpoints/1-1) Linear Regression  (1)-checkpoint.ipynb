# 학습 / 검증데이터 분할

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(raw_X, raw_y, train_size=0.7, test_size=0.3, random_state=1)
print(train_X.shape, test_X.shape, train_y.shape, test_y.shape)
```

```python
lin_model_5 = sm.OLS(train_y, train_X)
lin_model_5 = lin_model_5.fit()
```

```python
plt.plot(np.array(lin_model_5.predict(test_X)), label='pred_y')
plt.plot(np.array(test_y), label='true_y')
plt.legend()
plt.show()
```

# 성능평가

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
ridge = Ridge(alpha=.1, fit_intercept=True)
lasso = Lasso(alpha=.1, fit_intercept=True)

ridge.fit(train_X, train_y)
lasso.fit(train_X, train_y)
```

```python
print("R square score for multiple linear regression model : {:.2f}".format(r2_score(lin_model_5.predict(test_X), test_y)))
print("R square score for Ridge model : {:.2f}".format(r2_score(ridge.predict(test_X), test_y)))
print("R square score for Lasso model : {:.2f}".format(r2_score(lasso.predict(test_X), test_y)))
```

```python
print("MSE for multiple linear regression model : {:.2f}".format(mean_squared_error(lin_model_5.predict(test_X), test_y)))
print("MSE for Ridge model : {:.2f}".format(mean_squared_error(ridge.predict(test_X), test_y)))
print("MSE for Lasso model : {:.2f}".format(mean_squared_error(lasso.predict(test_X), test_y)))
```

```python
alpha_list = np.linspace(0.1, 100, 500)

ridge_mse_list_train = []
ridge_mse_list_test = []

lasso_mse_list_train = []
lasso_mse_list_test = []

for alpha in alpha_list:
    lasso = Lasso(alpha=alpha)
    ridge = Ridge(alpha=alpha)
    lasso = lasso.fit(train_X, train_y)
    ridge = ridge.fit(train_X, train_y)
    
    lasso_mse_list_train.append(mean_squared_error(lasso.predict(train_X), train_y))
    ridge_mse_list_train.append(mean_squared_error(ridge.predict(train_X), train_y))
    
    lasso_mse_list_test.append(mean_squared_error(lasso.predict(test_X), test_y))
    ridge_mse_list_test.append(mean_squared_error(ridge.predict(test_X), test_y))
    
plt.plot(alpha_list, lasso_mse_list_train, label='Lasso MSE Train')
plt.plot(alpha_list, ridge_mse_list_train, label='Ridge MSE Train')
plt.plot(alpha_list, lasso_mse_list_test, label='Lasso MSE Test')
plt.plot(alpha_list, ridge_mse_list_test, label='Ridge MSE Test')
plt.xlabel('alpha')
plt.ylabel('MSE')
plt.legend()
```

# K-Fold

```python
alpha_list = np.linspace(0.1, 100, 500)

ridge_mse_list_train = []
ridge_mse_list_test = []

lasso_mse_list_train = []
lasso_mse_list_test = []

for alpha in alpha_list:
    lasso = Lasso(alpha=alpha)
    ridge = Ridge(alpha=alpha)
    
    #kfold 통해 성능평가
    lasso_mse_list_train.append(-np.mean(cross_val_score(lasso,train_X,train_y,scoring='neg_mean_squared_error',cv=10)))
    ridge_mse_list_train.append(-np.mean(cross_val_score(ridge,train_X,train_y,scoring='neg_mean_squared_error',cv=10)))
    
    #학습
    lasso.fit(train_X,train_y)
    ridge.fit(train_X,train_y)
    
    #테스트
    lasso_mse_list_test.append(mean_squared_error(test_y,lasso.predict(test_X)))
    ridge_mse_list_test.append(mean_squared_error(test_y,ridge.predict(test_X)))

    
plt.plot(alpha_list, lasso_mse_list_train, label='Lasso MSE Train')
plt.plot(alpha_list, ridge_mse_list_train, label='Ridge MSE Train')
plt.plot(alpha_list, lasso_mse_list_test, label='Lasso MSE Test')
plt.plot(alpha_list, ridge_mse_list_test, label='Ridge MSE Test')
plt.legend()
```

- 전반적으로 K-Fold 했을 시 MSE 상승
- 학습데이터에 대해서 테스트 한번 했을 때보다 K-Fold 했을 시 성능을 좀 더 정확히 확인 가능
