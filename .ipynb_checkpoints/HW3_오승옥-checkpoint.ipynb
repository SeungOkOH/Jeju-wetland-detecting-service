```python
import numpy as np
import matplotlib.pyplot as plt

# 변수 정의 
# 크기가 8131x1051인 2차원 배열을 생성
L = np.zeros((8131, 1051))
es = np.zeros((8131, 1051))
theta = np.zeros((8131, 1051))
theta_e = np.zeros((8131, 1051))
r = np.zeros((8131, 1051))

# 주어진 범위의 x축과 y축을 0.1 마다 자름
t = np.linspace(-70, 35, 1051)
p = np.linspace(1013, 200, 8131)

# 2차원 배열로 바꾸기
T, P = np.meshgrid(t, p)

# 각 좌표별 변수 계산
for i in range(8131):
    for j in range(1051):
        theta[i, j] = (T[i, j] + 273) * (1000 / P[i, j]) ** (2/7)
        L[i, j] = - 0.0000614342 * T[i,j]**3 + 0.00158927 * (T[i,j]**2) - 2.36418 * T[i,j] + 2500.79
        L[i, j] = 1000 * L[i, j]
        es[i, j] = 6.11657 * np.exp(L[i, j] / 461.5 * (1 / 273 - 1 / (T[i, j] + 273)))
        r[i, j] = 0.622 * es[i, j] / (P[i, j] - es[i, j]) * 1000
        theta_e[i, j] = theta[i, j] * np.exp(L[i, j] * r[i, j] / (1004.5 * (T[i, j] + 273) * 1000))
```

```python
# plotting 

plt.figure(figsize=(16, 10))

# 등온위선 그리기
theta_line = plt.contour(T, P, theta, levels=np.arange(200, 500, 5), colors='b')
plt.clabel(theta_line, inline=True, fontsize=8, levels=np.arange(200, 400, 10))

# 등상당온위선 그리기
theta_e_line = plt.contour(T, P, theta_e, levels=np.arange(240, 400, 10), colors='r',linestyles='dotted')
plt.clabel(theta_e_line, inline=True, fontsize=8, levels=np.arange(240, 400, 10))

# 등혼합비선 그리기
r_line = plt.contour(T, P, r, levels=(0.1, 0.2, 0.5, 1, 1.5, 2, 3, 4, 6, 8, 10, 12, 15, 20, 25, 30), colors='lime')
plt.clabel(r_line, inline=True, fontsize=8, levels=(0.1, 0.2, 0.5, 1, 1.5, 2, 3, 4, 6, 8, 10, 12, 15, 20, 25, 30))


# y축 반전
plt.gca().invert_yaxis()

# y축 스케일을 p^(2/7)으로 수정해 건조단열선을 직선으로 만들기
plt.gca().set_yscale('function', functions=(lambda y: y**(2/7), lambda y: y**(7/2)))

plt.xlabel("Temperature (°C)")
plt.ylabel("Pressure (hPa)")


# 주피터에서 한글 지원 글씨체
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# 건조단열선 레이블
plt.text(0.95, 0.95, '건조단열선', color='blue', ha='right', va='top',fontsize=20, weight='bold', transform=plt.gca().transAxes)

# 습윤단열선 레이블
plt.text(0.95, 0.9, '습윤단열선', color='red', ha='right', va='top', fontsize=20, weight='bold', transform=plt.gca().transAxes)

# 포화혼합비선 레이블
plt.text(0.95, 0.85, '포화혼합비선', color='lime', ha='right', va='top', fontsize=20, weight='bold', transform=plt.gca().transAxes)

plt.show()
```
