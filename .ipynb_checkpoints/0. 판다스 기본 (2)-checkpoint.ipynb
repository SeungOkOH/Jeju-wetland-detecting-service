# Pandas

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

## 인덱싱

```python
#데이터 프레임 만들기 방법
```

```python
#딕셔너리 기반의 데이터프레임 생성
Pandas_data=pd.DataFrame({
    "1":['A',"B","A","B","A"],
    "2":[2.1, 4.7, 2.5, 4.5, 2.3],
    "3":[1500, 1600, 1700, 1800, 1900],
    "4":[12,22,32,25, 31]
})
```

```python
Pandas_data
```

```python
arr = np.array([['A',2.1, 1500, 12],
          ['B', 4.7, 1600, 22], 
          ['A', 2.5, 1700, 32],
          ['B', 4.5, 1800, 25],
          ['A', 2.3, 1900, 31]])

pd.DataFrame(arr, columns =['1', '2', '3', '4'])
```

####  loc  : 행이나 컬럼의 이름으로 불러옴  ex) loc[행 값, 열 값]
#### iloc : 행이나 컬럼의 순서로 불러옴    ex) iloc[행 인덱스, 열 인덱스]  

```python
#인덱스 이름을 통한 인덱싱 0부터 3
Pandas_data.loc[0:3]
```

```python
#인덱스 이름을 통한 인덱싱 0과 2
Pandas_data.loc[[0,2]]
```

```python
#인덱스 이름을 통한 인덱싱, 인덱스 0부터 2, 변수 1, 2
Pandas_data.loc[0:2,["1", '2']]
```

```python
#인덱스 번호를 통한 인덱싱, 인덱스 0부터 2, 변수 0번째, 2번째 
Pandas_data.iloc[0:3,[0,2]]
```

```python
#인덱스 번호를 통한 인덱싱, 인덱스 0부터 2, 변수 0번째, 2번째 
Pandas_data[0:3][["1", '2']]
```

```python
#인덱스 변경
Pandas_data2=Pandas_data.copy()
Pandas_data2.index=["A","D","C","B", "E"]
Pandas_data2
```

```python
#Index가 0~2가 아니므로.
Pandas_data2.loc[0:2,["1", '2']]
```

```python
#인덱스 번호를 통한 인덱싱, 인덱스 처음부터 3개, 변수 첫번째, 세번째
Pandas_data2.iloc[0:3,[0,2]]
```

```python
#인덱스 번호를 통한 인덱싱, 인덱스 처음부터 3개, 변수명 1, 2
Pandas_data2[0:3][["1", '2']]
```

## 특정 행 불러오기

```python
Pandas_data2
```

```python
#변수명 4가 30 초과 or 변수명 1이 A없는 데이터
Pandas_data2.loc[(Pandas_data2['4']>30)|(Pandas_data2['1']!='A')]
```

```python
#변수명 1에서 A나 C가 있는 데이터 
Pandas_data2.loc[Pandas_data2['1'].isin(['A', 'C'])]
```

```python
#object 타입이 아닌 변수들
Pandas_data2.loc[:,Pandas_data2.dtypes!='object']
```

```python
#첫번쨰 3번쨰 마지막 인덱스만 인덱싱
Pandas_data2.loc[[True, False, True, False, True]]
```

```python
#변수 2번쨰 마지막 인덱싱
Pandas_data2.loc[:,[False, True, False, True]]
```

# 데이터 수정

```python
Pandas_data2
```

```python
#변수명 4의 값에서 5를 추가 
Pandas_data2['4']=Pandas_data2['4'].apply(lambda x: x+5)
Pandas_data2
```

```python
#변수명 3의 값에서 5 나눔 
Pandas_data2['3']=Pandas_data2['3'].apply(lambda x: x/5)
Pandas_data2
```

```python
Pandas_data3=Pandas_data2.copy()
Pandas_data3

#새로운 변수명 5 생성 
Pandas_data3['5']=[10,20,30,40,50]
Pandas_data3
```

# 데이터 불러오기

```python
data1=pd.read_csv("data/Pandas_data.csv")
data2=pd.read_csv("data/Pandas_data2.csv")
data3=pd.read_csv("data/Pandas_data3.csv")
```

```python
data1
```

```python
data2
```

```python
data3
```

# merge

```python
#data1, data2를 변수명 1,2를 기준으로 교집합
dataset_merge = pd.merge(data1, data2, how='inner',on=['1','2'])
dataset_merge
```

# concat

```python
#data2에서, data3을 아래로 붙임
dataset_concat = pd.concat([data2,data3],axis=0)
dataset_concat
```

```python
#data2에서 data3을 우로 붙임
dataset_concat = pd.concat([data2,data3],axis=1)
dataset_concat
```

# 그룹화

```python
dataset_merge
```

```python
dataset_merge.groupby('1')
```

```python
#변수명 1 기준으로 묶어 평균 계산
dataset_merge.groupby('1').mean()
```

```python
#변수명 1 기준으로 묶어 분산 계산
dataset_merge.groupby('1').var()
```

```python
#변수명 1 기준으로 묶어 다양한 통계랑 계산(개수, 평균, 표준편차,최소,1분위수, 2분위수, 3분위수, 최대)
dataset_merge.groupby('1').describe()
```
