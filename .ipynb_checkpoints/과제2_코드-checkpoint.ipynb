## • Preprocessing

- 전처리는 각자의 아이디어를 반영하여 진행

````python
# 의학적으로 이상치 제거
rawdata = rawdata[rawdata['LDL'] < 500]
rawdata = rawdata[rawdata['AST'] < 500]
rawdata = rawdata[rawdata['ALT'] < 1000]
````

````python
x = rawdata.drop('smoking', axis=1)
y = rawdata['smoking']
````

## • Feature Selection by Dimensionality Reduction

- mRMR, SVM-RFE 등 차원축소 기법을 사용하여 변수 선택 진행
- **(중요)** 분석의 재현이 가능하도록 차원축소 기법으로 최종 선택한 변수명을 셀에 출력 (미출력시 감점 -5점)

````python
k = 10
S = []

corr_matrix = (pd.concat([x, y], axis=1)
               .corr()
               .drop('smoking'))

highest_corr_idx = np.argmax(corr_matrix['smoking'])
first_feature = corr_matrix.index[highest_corr_idx]

S.append(first_feature)

for i in range(k - 1):
    relevance = np.abs(corr_matrix['smoking'])
    redundancy = np.mean(np.abs(corr_matrix.loc[:, S]), axis=1)
    candidate_feature = (relevance - redundancy)
    candidate_feature = candidate_feature.drop(S)
    best_x_idx = np.argmax(candidate_feature)
    next_feature = candidate_feature.index[best_x_idx]
    S.append(next_feature)

# 선택된 특성 출력
print("Selected Features:", S)
````

### 1. Random Forest

### Cross Validation / Hyper-parameter Tuning
각자의 아이디어로 교차검증 및 하이퍼파라미터 최적화를 진행

````python
col_names = S
x = x[col_names]
y = rawdata['smoking']

submission_data = submission_data[col_names]
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, stratify=y, random_state=RANDOM_SEED)
x_trainset, x_val, y_trainset, y_val = train_test_split(x_train, y_train, test_size=0.2, shuffle=True, stratify=y_train, random_state=RANDOM_SEED)
````

````python
rf = RandomForestClassifier(random_state=RANDOM_SEED)
````

````python
rf_params = {
    'n_estimators': np.arange(100, 1000, 100),
    'max_depth': [1, 3, 7, 10],
    'max_leaf_nodes': [1, 3, 7, 10],
    'max_features': np.arange(0.2, 1.2, 0.2),
    'min_samples_split': np.arange(0.2, 1.2, 0.2),
    'min_samples_leaf': np.arange(0.1, 1.0, 0.2),
}
````

````python
rf = RandomForestClassifier()
rnd_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=rf_params,
    scoring='f1_weighted',
    n_iter=100,
    cv=StratifiedKFold(n_splits=10),
    random_state=RANDOM_SEED,
)
cv_res = rnd_search.fit(x_trainset, y_trainset)
````

### • Modeling

- 평가지표는 $F1 \ Score$로 함

**⚠️ 주의사항**:
> 모든 tree 기반의 모델 생성시 시드 설정 (regression, classification 모델 모두 동일)
>
> ```python
> # 예시
> # default RANDOM_SEED is 486
> rf = RandomForestClassifier(..., random_state=RANDOM_SEED)
> catboost = CatBoostClassifier(..., random_seed=RANDOM_SEED)
> ```

````python
import joblib
best_params = cv_res.best_params_

# 최적의 파라미터를 사용하여 새로운 RandomForestClassifier를 생성
best_rf = RandomForestClassifier(random_state=RANDOM_SEED, **best_params)

# 최적의 모델을 전체 훈련 세트에 fitting
best_rf.fit(x_train, y_train)

# 모델 저장
best_model_filename = "best_random_forest_model.joblib"
joblib.dump(best_rf, best_model_filename)

loaded_best_rf = joblib.load(best_model_filename)
````

````python
print('Best Hyperparameters:', cv_res.best_params_)
print('Best F1-Weighted:', cv_res.best_score_)
````

````python
# x_test에서 f1 구하기
pred = cv_res.best_estimator_.predict(x_test)
f1 = f1_score(y_test, pred, average='weighted')

print("F1 Score:", f1)
````

````python
# submission_data에 대한 예측
submission_pred = cv_res.best_estimator_.predict(submission_data)
````

````python
# submission_pred를 DataFrame으로 변환
submission_result = pd.DataFrame({
    'RF': submission_pred
})

# CSV 파일로 저장
submission_result.to_csv('RF.csv', index=False)
````

### Cross Validation / Hyper-parameter Tuning  & Modeling

각자의 아이디어로 교차검증 및 하이퍼파라미터 최적화를 진행


평가지표는 $F1 \ Score$로 함

````python
catb_params = {
    'learning_rate': [0.01, 0.05, 0.1],
    'iterations': np.arange(100, 700, 100),
    'max_depth': [3, 7, 10],
    'min_data_in_leaf': np.arange(0.3, 1.0, 0.2),
}
````

````python
from catboost import CatBoostClassifier, Pool, cv
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, f1_score
import numpy as np

train_pool = Pool(data=x_train, label=y_train)
val_pool = Pool(data=x_val, label=y_val)
test_pool = Pool(data=x_test, label=y_test)

catb = CatBoostClassifier(random_state=RANDOM_SEED, verbose=0)

f1_scorer = make_scorer(f1_score, average='weighted')

rnd_search = RandomizedSearchCV(estimator=catb, param_distributions=catb_params, n_iter=100, cv=10, scoring=f1_scorer, random_state=RANDOM_SEED)
rnd_search.fit(x_train, y_train, eval_set=(x_val, y_val))

# 최적의 파라미터, 모델 저장
best_params = rnd_search.best_params_
best_model = rnd_search.best_estimator_

# 최적의 파라미터 출력
print("Best Parameters:", best_params)

# 최적의 모델을 전체 훈련 세트 fitting
best_model.fit(x_train, y_train, eval_set=(x_val, y_val))

# x_test에서 f1 구하기
y_pred = best_model.predict(x_test)
test_f1_score = f1_score(y_test, y_pred, average='weighted')
print("Test F1 Score:", test_f1_score)
````
