```python
import numpy as np
from datetime import datetime, timedelta, date
import matplotlib.pyplot as plt

print("1번")
print('cloud fraction Hourly available solar energy at surface at LN[W/m^2]')
print('-'*len('cloud fraction'),'-'*len('Hourly available solar energy at surface at LN[W/m^2]'))
#기본정보
latitude = np.radians(37.57) #위도
S = 1367

#태양 declination과 (a/r)^2을 구하는 상수
an=[1.000110,0.034221,0.000719]
bn=[0,0.001280,0.000077]
cn=[0.006918,-0.399912,-0.006758,-0.002697]
dn=[0,0.070257,0.000907,0.000148]

sun = 0 #sun declination
n=0
t=2*np.pi*325/365 #11월 22일 t값

#sun declination 구하기
for k in range(0,4):
    sun = sun + cn[k] * np.cos(k * t) + dn[k] * np.sin(k * t)
    
#Half day length 11/22   
H = np.arccos(-np.tan(latitude)*np.tan(sun)) 

#(a/r)^2 구하기
ar2 = 0
for j in range(0, 3):
    ar2 = ar2 + an[j] * np.cos(j * t) + bn[j] * np.sin(j * t)
    
Q = []   
for cloud_fraction in range(0,11,1) : #cloud fraction 0부터 10까지 1간격으로 반복계산
    TOD = 0.29 + 0.151 * cloud_fraction #total optical depth
    Q_hour = S*ar2*(np.sin(latitude)*np.sin(sun) + 24/np.pi*np.cos(latitude)*np.cos(sun)*np.sin(np.pi/24)*np.cos(np.radians(0))) * np.exp(-TOD) #hourly solar energy at surface at LN
    print('{}              {}'.format(cloud_fraction, Q_hour))
    Q.append(Q_hour)
    
#그래프 그리기
x = np.arange(0,11)

plt.plot(x,Q)
plt.xlabel('cloud fraction')
plt.ylabel('Hourly available solar energy at surface at LN[W/m^2]')
plt.show()
```

```python
#2
print("2번")
print('elevation angle(˚) daily solar insolation[W/m^2]')
print('-'*len('elevation angle(˚) daily solar insolation[W/m^2]'))

sun = 0 #autumn equinox
n=0
t = 2*np.pi*265/365 #autumn equinox(9월 22일) t 값

for c, d in zip(cn,dn) : #sun declination
    sun += c*np.cos(n*t) +d*np.sin(n*t)
    n +=1
    
H = np.arccos(-np.tan(latitude)*np.tan(sun)) #half day length, equinox이기 때문에 위도에 상관없이 동일

n=0
ar2=0

for a,b in zip(an,bn) : #(a/r)^2
    ar2 += a*np.cos(n*t)+b*np.sin(n*t)
    n+=1
    
l = []
for B in range(0,91,5) : #0도부터 
    Q_daily = 1367/np.pi*ar2*(np.sin(latitude-np.radians(B))*np.sin(sun)*H + np.cos(latitude-np.radians(B))*np.cos(sun)*np.sin(H)) #daily solar insolation 계산식
    l. append(Q_daily)
    print('{}                            {}'.format(B,Q_daily))
    
print('*max daily solar insolation\nelevation angle(˚) : {}\ndaily solar insolation[W/m^2] : {}'.format(l.index(max(l))*5, max(l)))

#plotting
x= np.arange(0,91,5)
plt.plot(x,l)
plt.xlabel('elevation angle(˚)')
plt.ylabel('daily solar insolation[W/m^2]')
```
