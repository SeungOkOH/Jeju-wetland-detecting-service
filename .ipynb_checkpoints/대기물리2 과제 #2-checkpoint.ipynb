```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.1 #cm-1
```

```python
x = np.arange(0,0.2+0.01, 0.01) # kappa - kappa_0
L = a/(np.pi*(a**2+x**2)) # Lorentz profile
D = 1/(a*np.sqrt(np.pi))*np.exp(-(x/a)**2) #Doppler profile
```

```python
#voigt profile
x_prime = x 
convol = np.zeros(x.shape[0], dtype = float)
for x_p in x_prime :
    f_L = 1/(x_p**2+a**2)
    f_D = np.exp(-((x-x_p)**2)/a**2)
    convol = convol + f_L*f_D
convol = convol*0.01

V = convol*a/a/np.pi**(3/2)
```

```python
fig = plt.figure(facecolor='white') # figure 열기
ax = fig.add_subplot(1,1,1) # axes 추가
ax.plot(x,L, label = 'Lorentz Broadening')
ax.plot(x,D, label = 'Doppler Broadening')
ax.plot(x, V, label = 'Voigt Profile')
ax.set_xlabel('W-W0', fontsize = 20)
ax.set_ylabel('k(W-W0)', fontsize = 20) # k= absorption coefficient
plt.legend()
plt.savefig('C:/Users/82104/Desktop/대기물리2 과제/HW2_오승옥.png') # 그림 저장 : 괄호 안에 경로 및 파일 이름
```
